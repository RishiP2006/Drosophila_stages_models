```python
# 0) Mount Google Drive (Colab only)
from google.colab import drive
drive.mount('/content/drive')

# 1) Imports
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.metrics import accuracy_score, classification_report

from tensorflow.keras.models import load_model
from tensorflow.keras import layers, models
from tensorflow.keras.applications import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping

# 2) Paths
DATA_DIR = "/content/drive/MyDrive/Drosophila_stages"
SAVE_MODEL_PATH = "/content/drive/MyDrive/drosophila_inceptionv3_classifier.h5"

# 3) Data generators
trainGen = ImageDataGenerator(
    validation_split=0.15,
    preprocessing_function=preprocess_input,
    horizontal_flip=True,
    vertical_flip=True,
    rotation_range=20
)
testGen = ImageDataGenerator(preprocessing_function=preprocess_input)

# 4) Load train & validation from the single folder
trainGenerator = trainGen.flow_from_directory(
    DATA_DIR,
    target_size=(299, 299),
    batch_size=64,
    subset='training',
    class_mode='categorical',
    shuffle=True
)
validationGenerator = trainGen.flow_from_directory(
    DATA_DIR,
    target_size=(299, 299),
    batch_size=64,
    subset='validation',
    class_mode='categorical',
    shuffle=False
)

# 5) Detect classes
num_classes = trainGenerator.num_classes
CLASS_NAMES = list(trainGenerator.class_indices.keys())
print("Detected classes:", CLASS_NAMES)

# 6) Build the model
base_model = InceptionV3(include_top=False, weights='imagenet', input_shape=(299, 299, 3))
for layer in base_model.layers:
    layer.trainable = False

model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(num_classes, activation='softmax')
])

# Unfreeze last 30 layers
for layer in base_model.layers[-30:]:
    layer.trainable = True

model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

# 7) Callbacks
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-6, verbose=1)

early_stop = EarlyStopping(monitor='val_loss', patience=4, restore_best_weights=True, verbose=1)

# 8) Train
history = model.fit(
    trainGenerator,
    validation_data=validationGenerator,
    epochs=25,
    steps_per_epoch=len(trainGenerator),
    validation_steps=len(validationGenerator),
    callbacks=[lr_scheduler, early_stop]
)

# 9) Save model
model.save(SAVE_MODEL_PATH)
print(f"Model saved to: {SAVE_MODEL_PATH}")

# 10) Plot loss
plt.figure()
plt.plot(history.history['loss'],  label="Train Loss")
plt.plot(history.history['val_loss'], label="Val Loss")
plt.title("Loss Curve")
plt.xlabel("Epoch"); plt.ylabel("Loss")
plt.legend()
plt.savefig("LossVal_loss.png")
plt.show()

# 11) Plot accuracy
plt.figure()
plt.plot(history.history['accuracy'],  label="Train Acc")
plt.plot(history.history['val_accuracy'], label="Val Acc")
plt.title("Accuracy Curve")
plt.xlabel("Epoch"); plt.ylabel("Accuracy")
plt.legend()
plt.savefig("AccVal_acc.png")
plt.show()

# 12) Evaluate on validation set
loaded_model = load_model(SAVE_MODEL_PATH)
preds = np.argmax(loaded_model.predict(validationGenerator), axis=1)
true = validationGenerator.classes

print(f"Validation Accuracy: {accuracy_score(true, preds)*100:.2f}%")
print("Classification Report:\n",
      classification_report(true, preds, labels=np.arange(num_classes), target_names=CLASS_NAMES))

# 13) Confusion matrix
cm = tf.math.confusion_matrix(true, preds).numpy()
plt.figure(figsize=(10,8))
sns.heatmap(cm, annot=True, fmt='g', cmap='Reds',
            xticklabels=CLASS_NAMES, yticklabels=CLASS_NAMES)
plt.xlabel("Predicted"); plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.xticks(rotation=45); plt.tight_layout()
plt.savefig("Confusion_Matrix.png")
plt.show()
```
